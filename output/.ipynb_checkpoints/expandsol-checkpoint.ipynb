```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
fac=10
```

```python
mol='heh+'
basis='6-31g'
```

```python
fold = np.load('./final_'+mol+'_'+basis+'.npy')
```

```python
numsteps = fold.shape[0]
print(numsteps)
```

```python
fnew = np.repeat(fold,fac) + 1e-1*np.random.normal(size=fac*numsteps)
```

```python
dt = 0.1
tvec = dt*np.arange(numsteps)
tvec2 = (dt/fac)*np.arange(numsteps*fac)
plt.plot(tvec,fold,color='black')
plt.plot(tvec2,fnew,color='red')
plt.show()
```

```python
np.save('./guess_'+mol+'_'+basis+'.npy',fnew)
```
